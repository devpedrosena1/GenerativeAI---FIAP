**Tarefa:** 

### Implementar um modelo preditivo de classificação para prever a categoria do corte (coluna 'cut') do diamante com base nas dimensões (colunas'x', 'y' e 'z')

```python
# Separação de dados de entrada X (features, variáveis independentes) 
# e dados de saída y (target, variável dependente)
```

```python
# X = df[['x', 'y', 'z']]
X = df.iloc[:, -3:] # : ---> todos os registros, -3: ---> da antepenúltima até o final
X
```

```python
y = df['cut']
y
```

```python
# Separação de dados de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

### Escolha o seu modelo/algoritmo de classificação

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

# Árvore de decisão 
from sklearn.tree import DecisionTreeClassifier

# Random Forest
from sklearn.ensemble import RandomForestClassifier

# Regressão Logística
from sklearn.linear_model import LogisticRegression
```

```python
# Modelo KNN
modelo_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# Treinamento do modelo
modelo_knn.fit(X_train, y_train)
```

```python
# Previsões do modelo KNN
y_predict = modelo_knn.predict(X_test)
y_predict
```

```python
# Avaliando modelo com base na acurácia
# y_test.values
from sklearn.metrics import accuracy_score

acuracia = accuracy_score(y_test, y_predict)
print(f'A acurácia do modelo KNN: {acuracia * 100: .2f}%')
```
